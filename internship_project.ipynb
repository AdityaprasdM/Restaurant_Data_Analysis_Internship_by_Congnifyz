# Task-1.1: Determine the top three most common cuisines in the dataset

```python
cuisines_long = (
    df[["Restaurant ID", "Cuisines"]]
    .dropna(subset=["Cuisines"])
    .assign(Cuisines=lambda d: d["Cuisines"].str.split(","))
    .explode("Cuisines")
    .assign(Cuisine=lambda d: d["Cuisines"].str.strip())
    .drop(columns=["Cuisines"])
    .query("Cuisine != ''")
)

cuisines_unique = cuisines_long.drop_duplicates(["Restaurant ID", "Cuisine"])

n_restaurants = df["Restaurant ID"].nunique()

counts = cuisines_unique["Cuisine"].value_counts()
top3 = counts.head(3)

result = (
    pd.DataFrame({"Count": top3})
    .reset_index()
    .rename(columns={"index": "Cuisine"})
)

print("Top 3 cuisines by restaurant coverage:\n")
print(result.to_string(index=False))
```

# Task-1.2 : Calculate the percentage of restaurants that serve each of the top cuisines.

```python
percent = (top3 / n_restaurants * 100).round(2)
print(f"{percent}")
```

# Level3 Task-1.1 :  Analyze the text reviews to identify the most common positive and negative keywords.

```python
import nltk
from nltk.corpus import stopwords
from collections import Counter
import string
```

```python
nltk.download('stopwords')
```

```python
reviews = dff['Rating text'].dropna().astype(str)
```

```python
stop_words = set(stopwords.words('english'))
punct = set(string.punctuation)
```

```python
def clean_words(text):
    text = text.lower()
    words = [w for w in text.split() if w not in stop_words and w not in punct]
    return words

all_words = []
for r in reviews:
    all_words.extend(clean_words(r))
```

```python
positive_words = ["good", "great", "excellent", "amazing", "delicious", "tasty", "best", "love"]
negative_words = ["bad", "poor", "worst", "terrible", "slow", "cold", "average", "disappointing"]


positive_counts = Counter([w for w in all_words if w in positive_words])
negative_counts = Counter([w for w in all_words if w in negative_words])
```

```python
print("Top Positive Keywords:")
print(positive_counts)

print("\nTop Negative Keywords:")
print(negative_counts)
```

# Level3 Task-3.1:  Analyze if there is a relationship between the price range and the availability of online delivery and table booking

```python
unique = dff.drop_duplicates(subset=['Restaurant ID'])
```

```python
unique['Has Online delivery'] = unique['Has Online delivery'].astype(str).str.strip().str.title()
unique['Has Table booking'] = unique['Has Table booking'].astype(str).str.strip().str.title()
```

```python
online_delivery_relation = (
    unique.groupby('Price range')['Has Online delivery']
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
    .unstack()
    .fillna(0)
)
```

```python
table_booking_relation = (
    unique.groupby('Price range')['Has Table booking']
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
    .unstack()
    .fillna(0)
)
```

```python
print("\n Percentage of Restaurants Offering Online Delivery by Price Range:\n")
print(online_delivery_relation)

print("\n Percentage of Restaurants Offering Table Booking by Price Range:\n")
print(table_booking_relation)
```

# Level3 Task-3.2: Determine if higher-priced restaurants are more likely to offer these services.

```python
service_relation = (
    unique.groupby('Price range')[['Has Online delivery','Has Table booking']]
    .apply(lambda g: g.apply(lambda col: col.value_counts(normalize=True).get('Yes',0)))
    .mul(100)
    .round(2)
)
```
